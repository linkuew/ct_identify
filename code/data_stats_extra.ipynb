```python
import pandas as pd
import nltk
import statistics as st
from nltk.tokenize import word_tokenize
import numpy as np
import matplotlib.pyplot as plt
```

```python
nltk.download('punkt')
```

```python
def get_data(seed):
    testfile = "../data/test_" + seed + ".csv"
    trainfile = "../data/train_" + seed + ".csv"
    valfile = "../data/val_" + seed + ".csv"
    
    with open(testfile) as f:
        testdata = pd.read_csv(testfile)

    with open(trainfile) as f:
        traindata = pd.read_csv(trainfile)
    
    with open(valfile) as f:
        valdata = pd.read_csv(valfile)
    
    alldata = pd.concat([testdata, traindata, valdata], ignore_index=True)
    
    return alldata
```

```python
def get_avg_len(data, label):
    lens = []
    for _, entry in data.iterrows():
        if entry["label"] == label:
            lens.append(len(word_tokenize(entry["text"])))
        
    return lens, st.mean(lens), st.stdev(lens)
```

```python
bf = get_data("big.foot")
vc = get_data("vaccine")
cc = get_data("climate")
pg = get_data("pizzagate")
fe = get_data("flat.earth")
```

**Big Foot**

```python
bflens, bfmean, bfstdev = get_avg_len(bf, "conspiracy")
```

```python
print(bfmean, bfstdev)
```

```python
fig, ax = plt.subplots()

ax.hist(bflens, bins=50)
ax.set_ylim(0, 250)
ax.set_ylabel("Number of Documents")
ax.set_xlabel("Words per document")
ax.set_title("Histogram of " + "big foot" + " " + "conspiracy" + " documents by length")
```
